**Assignment No. :5** Generate creative artwork using Stable Diffusion

```python
from diffusers import StableDiffusionPipeline
import torch

def generate_image(prompt, model_name="CompVis/stable-diffusion-v1-4"):
    """
    Generate an image from a text prompt using a pre-trained model.
    """
    # Load pre-trained model
    pipe = StableDiffusionPipeline.from_pretrained(model_name)
    pipe.to(torch.device("cuda" if torch.cuda.is_available() else "cpu"))

    # Generate image
    image = pipe(prompt).images[0]

    # Save and display image
    image_path = "generated_image.png"
    image.save(image_path)
    return image_path
```

```python
# Example usage
if __name__ == "__main__":
    prompt = "A Bird on a tree"
    image_path = generate_image(prompt)
    print(f"Image saved at: {image_path}")
```
