**Assignment No. 4:** Build a simple question-answering chatbot using a pre-trained language model

**Steps:-**
1.Install & Import Libraries – Set up transformers and import necessary modules.

2.Load Pre-trained Model – Use distilbert-base-cased-distilled-squad, a lightweight version of BERT fine-tuned on SQuAD for question-answering.

3.Get Context – Ask the user to provide background information.

4.Take Questions – Continuously accept user queries.

5.Generate Answers – Use the DistilBERT model to extract answers from the provided context.

6.Exit Condition – Allow users to exit by typing "exit".

```python
import torch
from transformers import pipeline
```

```python
qa_pipeline = pipeline("question-answering", model="distilbert-base-cased-distilled-squad")
```

```python
def chatbot():
    print("Simple QA Chatbot (Type 'exit' to stop)")
    context = input("Provide some context:\n")  # User provides background info

    while True:
        question = input("\nAsk a question: ")
        if question.lower() == "exit":
            print("Goodbye!")
            break

        # Get the answer from the model
        answer = qa_pipeline(question=question, context=context)
        print(f"Answer: {answer['answer']}")
```

```python
# Run the chatbot
chatbot()
```
